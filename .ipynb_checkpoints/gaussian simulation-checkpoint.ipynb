```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('C:\\Users\\zzuge\\공시지가 예측.csv')
```

```python
df=df.drop(['Unnamed: 0', 'X'],axis=1)
```

```python
df
```

```python
df=df[df['demand'] <0]
df['demand']=-(df['demand'])
df['demand']=round(df['demand'])
df=df[df['demand']>0]
```

```python
df['pred.y']=10**df['pred.y']
```

```python
df['cost']=df['demand']*df['pred.y']
```

```python
total_demand=df['demand'].sum()
```

```python
total_demand
```

```python
df=df.sort_values('demand',ascending=False)
```

```python
df
```

```python
df['demand_cumsum']=df['demand'].cumsum()
```

```python
df['cost_cumsum']=df['cost'].cumsum()
```

```python
df
```

```python
df['demand_cumsum']=total_demand-df['demand_cumsum']
df
```

```python
df['idx']=range(0,len(df))
df=df.set_index('idx')
```

```python
df
```

```python
plt.figure(figsize=(15, 10))
fig, ax1 = plt.subplots()
ax1.plot(df['demand_cumsum'],label='demand',linewidth="5",color='red')
ax2 = ax1.twinx()
plt.plot(df['cost_cumsum'],label='supply',linewidth="5",color='blue')
plt.show()
```

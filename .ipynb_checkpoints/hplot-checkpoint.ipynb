```julia
using Plots

crd = data1[:, [:cell_xcrd, :cell_ycrd]]
rename!(crd, ["x", "y"])

h = 0.005
d = trunc(Int64, h |> log10 |> abs)
crd = trunc.(crd, digits = d)

X = crd.x
Y = crd.y
xmin = floor(minimum(X), digits = d)
ymin = floor(minimum(Y), digits = d) # 당현리
xmax =  ceil(maximum(X), digits = d)
ymax =  ceil(maximum(Y), digits = d) # 유사리

xsize = length(xmin:h:xmax)
ysize = length(ymin:h:ymax)
mesh = zeros(Int64, ysize, xsize)
for (x,y) in zip(X,Y) # 3.64 seconds
    mesh[
        findfirst(ymin:h:ymax .== y),
        findfirst(xmin:h:xmax .== x)
    ] += 1
end
sum(mesh) # check missing cell

heatmap(mesh, aspect_ratio = 1)
```

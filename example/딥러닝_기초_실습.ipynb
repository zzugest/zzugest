```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/소득train.csv')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df=df.drop('id',axis=1)
```

```python
df['target'].value_counts()
```

```python
df=pd.get_dummies(df)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
y_target = df['target']
x_features =df.drop(['target'],axis=1)
x_train, x_test, y_train, y_test = train_test_split(x_features, y_target, test_size = .3, random_state=777)
```

```python
x_train
```

```python
y_train
```

```python
from tensorflow import keras
```

```python
logistic = keras.Sequential() # 모델 생성
```

```python
logistic.add(keras.layers.Input(shape=(104))) # 독립변수의 개수
for i in range(10):  # 반복문을 이용해 레이어 add
    logistic.add(keras.layers.Dense(100)) 
    logistic.add(keras.layers.Activation('relu'))
```

```python
logistic.add(keras.layers.Dense(1)) # 종속변수의 개수
```

```python
logistic.add(keras.layers.Activation('sigmoid')) # 로지스틱이기 때문에 시그모이드 
```

```python
logistic.summary()
```

```python
logistic.compile(
    optimizer='adam', 
    loss='binary_crossentropy', 
    metrics=['acc']
)
```

```python
logistic.fit(x_train, y_train,epochs=1000)
```

```python
predict=logistic.predict(x_test)
```

```python
predict
```

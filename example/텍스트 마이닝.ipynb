```python
# 아래 코드는 konlpy라는 패키지 안에 tag 모듈 안에 있는 Okt라는 모듈을 임포트 합니다.
# konlpy 안에 tag 안에 Okt가 있는 이런 상하 관계는 누가 정했냐구요? konlpy 개발자가 그렇게 정했고,
# 저는 지금 konlpy 홈페이지에 적혀져 있는 사용법대로 사용하고 있는 겁니다.
# konlpy 홈페이지 : https://konlpy-ko.readthedocs.io/ko/v0.4.3/

# from 패키지.모듈 import 하위모듈
# 이것은 패키지를 임포트하는 전형적인 방식입니다.
```

```python
from konlpy.tag import Okt
```

```python
tokenizer = Okt()
```

```python
kor_text = '밤에 귀가하던 여성에게 범죄를 시도한 대 남성이 구속됐다서울 제주경찰서는 \
            상해 혐의로 씨를 구속해 수사하고 있다고 일 밝혔다씨는 지난달 일 피해 여성을 \
            인근 지하철 역에서부터 따라가 폭행을 시도하려다가 도망간 혐의를 받는다피해 \
            여성이 저항하자 놀란 씨는 도망갔으며 신고를 받고 주변을 수색하던 경찰에 \
            체포됐다피해 여성은 이 과정에서 경미한 부상을 입은 것으로 전해졌다'

# 문장에서 명사만 추출 (100% 정확하진 않아요!)
print(tokenizer.nouns(kor_text))
```

입력된 문자열이 명사 단위로 토큰화 된 것을 볼 수 있습니다. 자연어 처리에서는 토큰화 결과가 얼마나 정확한지에 따라서 분석 결과에 영향을 주기도 하므로 형태소 분석기의 성능 또한 텍스트 데이터 분석에 굉장히 중요합니다.

이렇게 토큰화를 하였다면 추가적으로 해야 하는 전처리가 있습니다. 바로 텍스트 분석에서 불필요한 토큰들을 제거하는 불용어 제거입니다. **불용어(stopwords)**란, 데이터 전체에서 꽤 많이 등장하지만 실제로는 별로 중요하지 않은 단어들을 말합니다.  한국어에서는 주로 ~가, ~은, ~는과 같은 조사나 접사 등이 불용어에 속합니다.  또는 해당 데이터의 특성으로 인해 자주 등장할 수 밖에 없는 단어들도 불용어입니다.

불용어는 미리 짐작하여 한 번에 정의하는 것이 아니라, 토큰화 과정을 거친 결과를 지속적으로 확인하면서 계속해서 추가하게 되는 것이 일반적입니다.

뉴스 기사 데이터이다보니 이 기사를 불펌하지 말라는 의미에서 '무단 배포 금지' 와 같은 문장이 계속해서 등장합니다. '무단 배포 금지'라는 문장은 뉴스 기사랑 상관없는 내용이므로 저는 이 문장이 뉴스 데이터의 내용에 어떤 내용들이 들어가있는지를 분석하기에 불필요하다고 생각했습니다. 그리고 이 '무단 배포 금지'라는 문장은 토큰화 과정에서 확인해본 결과 주로 '무단', '배포', '금지' 라고 토큰화 되는 경우가 많았습니다. 그래서 불용어 리스트에다가 추가해줬습니다.

이제 뉴스 기사 데이터를 토큰화하는 방법을 알아봅시다.

- **데이터프레임이름['열의 이름'].apply(동작 코드)**

를 사용하면 **동작 코드를 데이터프레임의 특정 열(column. 즉, 세로 방향) 전체. 다시 말해 해당 열에 전부 적용하는 효과를 가집니다.**

```python
'''
데이터프레임이름['열의 이름'].apply(어떤 함수)를 사용하면
어떤 함수를 데이터프레임의 특정열 전체에 전부 적용하는 효과를 가집니다.
즉, 이 경우 데이터프레임의 news열에 tokenized.nouns라는 함수를 모두 적용하고 나서 tokenized 열에 새로 저장하라!
라는 의미가 됩니다.
'''
df['tokenized'] = df['news'].apply(tokenizer.nouns)
```

위의 경우 데이터프레임의 news열에 tokenized.nouns라는 함수를 모두 적용하고 나서 tokenized 열에 새로 저장하라! 라는 의미가 됩니다.

```python
df
```

```python
stop_words = ['기자', '제공', '무단', '배포', '무단배포', '배포금지', '이번', '위해', '라며', '금지', '뉴스', '통해', '오늘', '지난달', '지난', '대한', '경우', '관련', '뉴시스', '현재', '지난해', '때문', '지금', '또한', '만큼', '최근', '당시', '올해', '대해', '다시', '모두']
```

리스트 컴프리헨션이라는 연산에 대해서 예시를 통해서 설명해봅시다.

```python
test_list = ['경찰서', '상해', '혐의', '씨', '구속', '수사', '일']
```

```python
remove_word_list = ['경찰서', '구속']
```

```python
test_list = [item for item in test_list if item not in remove_word_list and len(item) > 1]
```

```python
test_list
```

위 코드는 전형적인 `리스트 컴프리헨션 연산`입니다.

파이썬에서 리스트를 사용할 때 굉장히 많이 사용되는 문법으로 익숙해지시는 게 좋습니다.

- **[item for item in test_list if item not in stop_words and len(item) > 1]**

**1) for item in test_list :**

test_list라는 파이썬 리스트 자료형에서 원소를 하나씩 꺼내는데 이를 item이라 명명

**2) if item not in remove_word_list and len(item) > 1**

item이 remove_word_list에 없으면서 이 item의 길이가 1초과인 경우를 조건으로 한다.

ex) '용' 이라는 단어는 길이가 1. 하지만 '학교'라는 단어는 길이가 2.

**3) [item for item in test_list if item not in remove_word_list and len(item) > 1]**

조건에 맞는 경우에만 **item을 남긴다.**

다시 말해 조건에 맞지 않으면 살아남지 못하여 없어지게 됩니다.

결과를 보면 remove_word에 존재하는 '경찰서'라는 단어와 '구속'이라는 단어가 제거되었고, 길이가 1이었던 '씨'라는 단어와 '일'이라는 단어가 제거된 것을 볼 수 있습니다.

```python
df['tokenized'] = df['tokenized'].apply(lambda x: [item for item in x if item not in stop_words and len(item) > 1])
```

- **df['tokenized'] = df['tokenized'].apply(lambda x: [item for item in x if item not in stop_words and len(item) > 1])**

위의 코드가 의미하는 바를 다시 해석해봅시다.

- lambda x : 동작 코드

는 현재 입력을 x라고 하고 x에 대해서 동작 코드를 수행하라 라는 의미입니다.

**본래 특정 데이터프레임의 열을 리스트로 만드는 코드**는 다음과 같습니다.

`df['특정열'].values`

이에 따라서

`df[df['code']=='특정 카테고리']['tokenized'].values`를 하면

특정 카테고리의 뉴스들의 데이터만을 리스트로 만들 수 있습니다. 

위 코드에서 밑줄 친 부분이 하나의 데이터프레임의 특정열에 해당되기 때문입니다.

정확히는 이중 리스트가 되겠네요. 하나의 리스트 안에 각 원소는 리스트가 될 것입니다.

```python
# 넘파이라는 패키지를 np라는 이름으로 임포트(사용 선언)
import numpy as np
```

np.hstack()은 여러 데이터를 하나의 데이터로 묶어서 만들어주는 함수입니다.

그래서 이중 리스트를 입력으로 하면, 하나의 리스트로 바꿔줍니다.

```python
# code 열의 값이 '사회'인 경우만 필터링한 데이터프레임의 'tokenized'열의 모든 리스트를 전부 연결하여 하나의 리스트로 변환.
social_news = np.hstack(df[df['code']=='사회']['tokenized'].values)

# code 열의 값이 '생활/문화'인 경우만 필터링한 데이터프레임의 'tokenized'열의 모든 리스트를 전부 연결하여 하나의 리스트로 변환.
life_news = np.hstack(df[df['code']=='생활/문화']['tokenized'].values)

# code 열의 값이 'IT/과학'인 경우만 필터링한 데이터프레임의 'tokenized'열의 모든 리스트를 전부 연결하여 하나의 리스트로 변환.
it_news = np.hstack(df[df['code']=='IT/과학']['tokenized'].values)
```

```python
# 하나의 리스트로 변환했다면 각각의 길이는? 매우 깁니다!
print(len(social_news))
print(len(life_news))
print(len(it_news))
```

단어의 등장 빈도를 카운트하는 아주 유용한 함수로 `collections`라는 패키지의 `Counter`라는 함수가 있습니다. 이해를 위해 테스트 리스트를 만들어 Counter가 수행해주는 일이 어떤 일인지 파악해봅시다.

```python
from collections import Counter
```

```python
test_list = ['바나나', '바나나', '바나나', '사과', '사과', '딸기', '메론', '수박', '수박', '토마토', '토마토', '전화기', '전화기', '전화기', '전화기']
```

```python
test_word_count = Counter(test_list)
print(test_word_count)
print(test_word_count.most_common(3))
```

등장 빈도를 카운트한 결과에 most_common(특정 숫자)을 사용할 경우에는 **입력한 특정 숫자를 순위로 인식**해서 등장 빈도수가 많은 순서를 높은 순위로 하였을 때, 입력한 숫자만큼의 순위만 출력합니다. 예를 들어서 most_common(3)을 하면 3등까지만 출력해줍니다.

앞서 배운 Counter 함수를 앞서 하나의 리스트로 만들었던 사회 뉴스에 대해 적용해보고, 그 중 상위 20개의 단어만 출력해봅시다.

```python
social_news_word_count = Counter(social_news)
print(social_news_word_count.most_common(20))
```

앞서 배운 Counter 함수를 앞서 하나의 리스트로 만들었던 생활/문화 뉴스에 대해 적용해보고, 그 중 상위 20개의 단어만 출력해봅시다.

```python
life_news_word_count = Counter(life_news)
print(life_news_word_count.most_common(20))
```

앞서 배운 Counter 함수를 앞서 하나의 리스트로 만들었던 생활/문화 뉴스에 대해 적용해보고, 그 중 상위 20개의 단어만 출력해봅시다.

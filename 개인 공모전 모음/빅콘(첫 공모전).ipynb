```python
from sklearn.metrics import mean_squared_error

def rmsle(y, pred):
    log_y = np.log1p(y)
    log_pred = np.log1p(pred)
    squared_error = (log_y - log_pred) ** 2
    rmsle = np.sqrt(np.mean(squared_error))
    return rmsle

def rmse(y, pred):
    return np.sqrt(mean_squared_error(y, pred,  sample_weight=None, multioutput='uniform_average', squared=True))
```

```python
# RMSLE, RMSE 계산
def calculate_model_score(y, pred):
    rmsle_value = rmsle(y, pred)
    rmse_value = rmse(y, pred)
    print('RMSLE: {0:.3f}, RMSE: {1:.5f}'.format(rmsle_value, rmse_value))
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
poly = PolynomialFeatures(degree=3, include_bias=False)  
y_target = df1['유입량']
x_features = df1.drop(['유입량','유평'], axis=1, inplace=False)

x_train, x_test, y_train, y_test = train_test_split(x_features, y_target, test_size = .3, random_state=777)


pr = LinearRegression()

x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.fit_transform(x_test)

pr.fit(x_train_poly, y_train)

y_train_pred = pr.predict(x_train_poly)
y_test_pred = pr.predict(x_test_poly)

print('훈련 RMSE: %.3f, 테스트 RMSE: %.3f' % (
        np.sqrt(mean_squared_error(y_train, y_train_pred)),
        np.sqrt(mean_squared_error(y_test, y_test_pred))))

print('훈련 R^2: %.3f, 테스트 R^2: %.3f' % (
        r2_score(y_train, y_train_pred),
        r2_score(y_test, y_test_pred)))
plt.plot(np.array(y_test_pred),label = "pred")
plt.plot(np.array(y_test),label = "true")
plt.legend()
plt.show()

```

```python
#이상치 제거는 맨처음 선형회귀할때 이용 
def remove_outlier(df, column):
    fraud = df[column]
    quantile_25 = np.percentile(fraud.values, 25)
    quantile_75 = np.percentile(fraud.values, 75)

    # IQR에 1.5를 곱한다.
    iqr = quantile_75 - quantile_25
    iqr_weight = iqr * 1.5

    # 최댓갑과 최솟값 연산.
    lowest_val = quantile_25 - iqr_weight
    highest_val = quantile_75 + iqr_weight

    # 아웃라이어 필터링 후 아웃라이어에 해당되는 인덱스 리턴
    outlier_index = fraud[(fraud < lowest_val) | (fraud > highest_val)].index
    return outlier_index
```

```python
outlier_index = remove_outlier(df=df1, column='A')
print('이상치 샘플의 인덱스:', outlier_index)
```
